```python
import requests
import pandas as pd

def fetch_news_data(api_key, query, start_date, end_date):
    url = f'https://newsapi.org/v2/everything'
    params = {
        'q': query,
        'from': start_date,
        'to': end_date,
        'apiKey': api_key,
        'language': 'en',
        'pageSize': 100  # Adjust based on needs
    }
    response = requests.get(url, params=params)
    data = response.json()

    # Check if the API call was successful and contains articles
    if data.get('status') == 'ok' and data.get('articles'):
        articles = data.get('articles', [])
        news_data = pd.DataFrame(articles)
        # Check if 'publishedAt' and 'title' are present in the articles
        if 'publishedAt' in news_data.columns and 'title' in news_data.columns:
            return news_data[['publishedAt', 'title']]
        else:
            print("Error: 'publishedAt' or 'title' not found in articles.")
            return None
    else:
        print(f"Error fetching news data: {data.get('message', 'Unknown error')}")
        return None

api_key = '9629020cba164297b08ee9c6cbfbb11b'  # Replace with your NewsAPI key
query = 'Apple Inc OR AAPL'
start_date = '2024-06-30'
end_date = '2024-07-29'
news_data = fetch_news_data(api_key, query, start_date, end_date)

if news_data is not None:  # Proceed only if news data was fetched successfully
    news_data.rename(columns={'publishedAt': 'Date', 'title': 'Headline'}, inplace=True)
    news_data['Date'] = pd.to_datetime(news_data['Date'])
    news_data.to_csv('news_headlines.csv', index=False)
    print("News data saved to 'news_headlines.csv'.")
```

```python
def preprocess_stock_data(stock_data):
    # Initial checks
    print("Initial data size:", stock_data.shape)

    # Ensure 'Adj Close' column exists
    if 'Adj Close' not in stock_data.columns:
        raise ValueError("'Adj Close' column is missing from the stock data")

    # Calculate returns
    stock_data['Return'] = stock_data['Adj Close'].pct_change()

    # Use a smaller window size if the dataset is small
    window_size = 20 if len(stock_data) > 20 else len(stock_data) - 1

    # Calculate rolling mean and standard deviation
    stock_data['Mean'] = stock_data['Adj Close'].rolling(window=window_size).mean()
    stock_data['StdDev'] = stock_data['Adj Close'].rolling(window=window_size).std()

    # Drop rows with NaN values
    stock_data.dropna(inplace=True)

    # Final check
    print("Final data size:", stock_data.shape)
    return stock_data

# Ensure there is sufficient data before preprocessing
if len(stock_data) < 20:
    print("Not enough data to perform rolling window calculations")
else:
    stock_data = preprocess_stock_data(stock_data)
```

```python
def generate_signals(data, threshold=1):
    data['Signal'] = 0
    data['Signal'][data['Adj Close'] < (data['Mean'] - threshold * data['StdDev'])] = 1  # Buy signal
    data['Signal'][data['Adj Close'] > (data['Mean'] + threshold * data['StdDev'])] = -1  # Sell signal
    return data

combined_data = generate_signals(combined_data)
```

```python
def remove_duplicate_dates(data):
    # Drop duplicate dates if they exist
    data = data[~data.index.duplicated(keep='first')]
    return data

def backtest_strategy(data, initial_cash=100000, commission=0.001):
    # Ensure there are no duplicate dates
    data = remove_duplicate_dates(data)

    cash = initial_cash
    shares = 0
    portfolio_values = []

    for i in range(len(data)):
        # Check the signal
        if data['Signal'].iloc[i] == 1:  # Buy signal
            if cash > 0:  # Ensure we have cash to buy
                shares = cash / data['Adj Close'].iloc[i]
                cash = 0  # All cash is used to buy shares
        elif data['Signal'].iloc[i] == -1 and shares > 0:  # Sell signal
            cash = shares * data['Adj Close'].iloc[i] * (1 - commission)
            shares = 0  # All shares are sold

        # Calculate the portfolio value
        portfolio_value = cash + shares * data['Adj Close'].iloc[i]
        portfolio_values.append(portfolio_value)

    # Add the final portfolio value to the data
    data['Portfolio Value'] = pd.Series(portfolio_values, index=data.index)
    return data

# Assuming combined_data has 'Adj Close' and 'Signal' columns
results = backtest_strategy(combined_data)

```

```python
import matplotlib.pyplot as plt

def evaluate_performance(results):
    initial_value = results['Portfolio Value'].iloc[0]
    final_value = results['Portfolio Value'].iloc[-1]
    total_return = (final_value - initial_value) / initial_value
    annualized_return = (1 + total_return) ** (252 / len(results)) - 1  # 252 trading days
    plt.figure(figsize=(12, 6))
    plt.plot(results.index, results['Portfolio Value'], label='Portfolio Value')
    plt.title(f'Algorithmic Trading Strategy Performance\nTotal Return: {total_return:.2%}, Annualized Return: {annualized_return:.2%}')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value ($)')
    plt.legend()
    plt.show()

evaluate_performance(results)
```
